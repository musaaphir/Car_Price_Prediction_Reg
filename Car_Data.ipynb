```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df["transmission"]=le.fit_transform(df["transmission"])
df["fuelType"]=le.fit_transform(df["fuelType"])
```

```python
df.head(2)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(feature):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = feature.columns
    vif["VIF"] = [variance_inflation_factor(feature.values, i) for i in range(feature.shape[1])]

    return(vif)
```

```python
feature = df.iloc[:,:-1]
calc_vif(feature)
```

```python
x=df.drop("log_price",axis=1,inplace=False)
x.head(2)
```

```python
y=df.log_price
y.head()
```

```python
import statsmodels.api as sm
```

```python
x_s=sm.add_constant(x)
slr=sm.OLS(y,x_s)
slr_model=slr.fit()
slr_model.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

## Linear Reg

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print('Variance score: {}'.format(lr.score(x_test, y_test)))
```

```python
print("Intercept :",lr.intercept_)
```

```python
print('Coefficients: ', lr.coef_)
```

```python
y_hat_lr=lr.predict(x_test)
y_hat_lr
```

```python
plt.figure(figsize=(16,4))
plt.plot(y_test,color="blue")
plt.plot(y_hat_lr,color="green")
plt.show()
```

```python
## plotting residual errors in training data
plt.figure(figsize=(15,6))
plt.scatter(lr.predict(x_train), lr.predict(x_train) - y_train, color = "green", s = 10, label = 'Train data')
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
plt.title("Residual errors Train Data")
plt.show()
```

```python
## plotting residual errors in test data
plt.figure(figsize=(15,6))
plt.scatter(lr.predict(x_test), lr.predict(x_test) - y_test,color = "blue", s = 10, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
plt.title("Residual errors test Data")
plt.show()
```

```python
## Combining both the Graph
plt.figure(figsize=(15,6))
plt.scatter(lr.predict(x_train), lr.predict(x_train) - y_train, color = "green", s = 10, label = 'Train data')
plt.scatter(lr.predict(x_test), lr.predict(x_test) - y_test,color = "blue", s = 10, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
plt.title("Residual errors")
plt.show()
```
